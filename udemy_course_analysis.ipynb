```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv('udemy_courses.csv')
```

# Minimum,Maximum,Average Course Duration

```python
avg_dur=data['content_duration'].mean()
max_dur=data['content_duration'].max()
min_dur=data['content_duration'].min()
data_min_dur=data.loc[data['content_duration']==min_dur]
data_max_dur=data.loc[data['content_duration']==max_dur]
course_min_dur=data_min_dur['course_title']
course_max_dur=data_max_dur['course_title']
print_data=f'Average Duration :{avg_dur}\nMaximum Duration :{max_dur}({course_max_dur})\nMinimum Duration :{min_dur}({course_min_dur})'
print(print_data)
```

# Which  Subject Highest reviews

```python
sb.scatterplot(x='subject',y='num_reviews',data=data)
```

Answer : Web Development

# Which Level of Highest and Lowest Subscribers

```python
sb.relplot(x="level", y="num_subscribers", hue="level",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=data)
```

Answer: Highest -Begginer Level
        lowest - Expert Level

# Which Subject of Max price in which Level

```python
sb.set(style="darkgrid", color_codes=True)
sb.stripplot(x="subject", y="price", hue='level',
              data=data, dodge=True, alpha=.25, zorder=1)
```

Answer: web development in Begginner level
